# Transformation Logic

```python
query = """
SELECT 
  cp.product_id,
  cp.product_number,
  cp.product_name,
  cp.category_id,
  pc.subcategory,
  pc.maintenance_flag,
  cp.product_line,
  cp.start_date
FROM silver.crm_products cp
LEFT JOIN silver.erp_product_category pc
  ON cp.category_id = pc.category_id 
-- WHERE cp.end_date IS NULL; --to get rid of historical data
"""
df = spark.sql(query)
```

```python
df.limit(10).display()
```

# Writing to Gold Layer

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.dim_products")
```
